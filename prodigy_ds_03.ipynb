```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Distribution of Accident severity
df['Accident_severity'].value_counts()
```

```python
#plotting the final class
sns.countplot(x = df['Accident_severity'])
plt.title('Distribution of Accident severity')
```

Visualize

```python
plt.figure(figsize=(10, 6))
pivot_table = df.pivot_table(index='Types_of_Junction', columns='Accident_severity', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='coolwarm', cbar=True)
plt.title("Heatmap of Accident Severity by Types of Junction")
plt.xticks(rotation=45)
plt.show()
```

Analyze patterns:

```python
# Calculate the count of accidents for each road surface type
road_surface_counts = df['Road_surface_type'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(road_surface_counts, labels=road_surface_counts.index, autopct='%1.1f%%', startangle=180)
plt.title("Distribution of Accidents by Road Surface Type")
plt.show()
```

```python
# Calculate the count of accidents for each road surface type
light_conditions_counts = df['Light_conditions'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(light_conditions_counts, labels=light_conditions_counts.index, autopct='%1.1f%%', startangle=180)
plt.title("Distribution of Accidents by Light conditions")
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Weather_conditions')
plt.title("Accidents by Weather Conditions")
plt.xticks(rotation=45)
plt.show()
     
```

```python
#Contributing Factors to Accidents
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Cause_of_accident')
plt.title("Contributing Factors to Accidents")
plt.xticks(rotation=90)
plt.show()
```
